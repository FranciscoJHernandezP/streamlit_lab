```python
import pandas as pd
import plotly.express as px
df_ss = pd.read_csv("https://raw.githubusercontent.com/jeaggo/tc3068/master/Superstore.csv")
df_states = pd.read_csv("https://raw.githubusercontent.com/jeaggo/datasets/master/States_USA.csv")
```

```python
df_states
```

```python
df = pd.merge(df_ss, df_states, on="State")
df.head()
```

```python
px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    hover_name="State",
    size="Quantity",
).update_layout(mapbox={"style": "carto-positron", "zoom": 4}, margin={"t":0,"b":0,"l":0,"r":0})
```

```python
fig = px.choropleth(df, locations="Abbreviation", locationmode="USA-states", color="Quantity", scope="usa")
fig.show()
```
